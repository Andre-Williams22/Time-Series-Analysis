```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("Data/LMND.csv")

df.tail()
```

```python
dates = df['Date']
opens = df['Open']
closes = df['Close']
highs = df['High']
lows = df['Low']
```

```python
plt.plot(dates, closes)
```

```python
plt.plot(dates, highs)
plt.plot(dates, lows)
plt.show()
```

```python
fig = plt.figure(figsize=(20, 10))
plt.subplot(221)
plt.plot(dates, opens)
plt.subplot(222)
plt.plot(dates, closes)
plt.subplot(223)
plt.plot(dates, highs)
plt.subplot(224)
plt.plot(dates, lows)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2)
axes[0, 0].plot(dates, opens)
axes[0, 1].plot(dates, closes)
axes[1, 0].plot(dates, highs)
axes[1, 1].plot(dates, lows)
plt.show()
```

```python
plt.plot(dates, closes)
plt.title('Close Prices of Lemonade')
plt.xlabel('Dates')
plt.ylabel('Prices in USD')
plt.xticks(pd.date_range('2020-07-29', '2020-10-28', 5).strftime('%Y-%m-%d'))
plt.show()
```
