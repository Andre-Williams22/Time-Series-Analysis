```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("LMND.csv")

df.tail()
```

```python
dates = df['Date']
opens = df['Open']
closes = df['Close']
highs = df['High']
lows = df['Low']
```

```python
plt.plot(dates, closes)
```

```python
plt.plot(dates, highs)
plt.plot(dates, lows)
plt.show()
```

```python
fig = plt.figure(figsize=(20, 10))
plt.subplot(221)
plt.plot(dates, opens)
plt.subplot(222)
plt.plot(dates, closes)
plt.subplot(223)
plt.plot(dates, highs)
plt.subplot(224)
plt.plot(dates, lows)
plt.show()
```
